# The most criminal part of Seattle is its downtown, but not for thefts of vechicles

## Geographic distribution of crimes in Seattle

Analysis of distribution crime incidents by the area of city and time is important for optimization work of police and offense preventions. 

As indicated on the map of crime density in Seattle, the highest rate of crime incidents per area is in downtown. Beat M3 has the least area, but the highest rate of crime (more than 1200 incidents in month per one square kilometer). Beats M2, M1, K1 are situated around the M3 and have very high rate of crime too. South and nothern parts of the city are much quiet. Incidents rate here usually not more than 100 in month per one square kilometer.

Theft of vechicles is the second most common type of crime in Seattle after the car prowls. It is not repeat geographic structure of total crime incidents. Places of vechicles thefts marked with blue dots on the map. They are about evenly distributed on the area of the city. Nothern part of the Seattle has a little bit higher density of this type of crime. South-Western part is the most secure from theft of vechicles perspective.

```python
%matplotlib inline
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import geopandas as gpd
from shapely.geometry import Point
#import contextily as ctx

seattle_data=pd.read_csv('seattle_incidents_summer_2014.csv', parse_dates=['Date Reported', 'Occurred Date or Date Range Start', 'Occurred Date Range End'], infer_datetime_format=True, low_memory=False) 
shpp = 'Seattle_Police_Beats_2018Present.shp'
```

```python
timeind = pd.DatetimeIndex(seattle_data['Occurred Date or Date Range Start'])
seattle_data['start_date'] = timeind.date.astype('datetime64')
seattle_data['start_hour'] = timeind.hour
seattle_data['day_of_week'] = timeind.dayofweek
```

```python
seattle_data['Offense Type'].value_counts()[:5]
```

```python
map_df = gpd.read_file(shpp)
map_df=map_df.to_crs(epsg=4326)
by_sector = seattle_data.groupby('Zone/Beat').count()
merged = map_df.set_index('beat').join(by_sector)
merged['density'] = (1.076e+7)*merged['RMS CDW ID']/merged['shape_Area']/3
vmin, vmax = 0, 1300
fig, ax = plt.subplots(1, figsize=(30, 20))
ax.set_title('Density of Criminal Incidents in Beats/Zones of Seattle (inc./(sq.km*month))', fontdict={'fontsize': 20, 'fontweight':3})
sm = plt.cm.ScalarMappable(cmap='OrRd', norm=plt.Normalize(vmin=vmin, vmax=vmax))
sm._A = []
cbar = fig.colorbar(sm)
variable='density'
merged['coords'] = merged['geometry'].apply(lambda x: x.representative_point().coords[:])
merged['coords'] = [coords[0] for coords in merged['coords']]
for idx, row in merged.iterrows():
    ax.annotate(s=idx, xy=row['coords'],
        horizontalalignment='center')
merged.plot(column=variable, cmap='OrRd', linewidth=0.8, ax=ax, edgecolor='0.8', vmin=vmin, vmax=vmax)

geo_seattle_data = seattle_data[seattle_data['Longitude'] != 0.0]
geo_seattle_data = geo_seattle_data[geo_seattle_data['Offense Type']=='VEH-THEFT-AUTO']
seattle_geometry = [Point(xy) for xy in zip(geo_seattle_data['Longitude'], geo_seattle_data['Latitude'])]
crs = {'init': 'epsg:4326'}
gdf = gpd.GeoDataFrame(geo_seattle_data, crs=crs, geometry=seattle_geometry)
gdf.plot(marker='o', markersize=3, color='b', ax=ax, legend=True, label='Motor vehicle theft')

fig.savefig("map_density.png", dpi=300)
```

## Dynamics of vechicle thefts

During summer months dynamics of vechicle thefts in Siettle has cyclic structure. Number of the thefts was slowly increasing from begining of all months to their ends. At the ends of the months, number of the incidents was sharply reduced. There is a general trend to reduce the number of thefts over the summer 2014 in Seattle.

```python
pivot_by_date = seattle_data.pivot_table('RMS CDW ID', aggfunc='count', index='start_date', columns='Offense Type')
pivot_by_date.fillna(0.0, inplace=True)
plt.figure(figsize=(10, 6))
pivot_by_date['ave_VEH-THEFT-AUTO']=pivot_by_date['VEH-THEFT-AUTO'].rolling(10, center=True).mean()
pivot_by_date['ave_VEH-THEFT-AUTO'].plot(color='r', linewidth=3, legend=True, label='10-days moving average')
pivot_by_date['VEH-THEFT-AUTO'].plot(linewidth=0.5, legend=True, label='Theft by day')
plt.title('Dynamics of vechicle thefts in Seattle (inc./day)', fontdict={'fontsize': 20, 'fontweight':3})
plt.show()
```

## Distribution of crime by day of week in different districts of Seattle

There are districts with increasing number of crime incidents in weekends in Seattle. For example, total number of crime incresed almost on 50% in district E from Friday to Sunday in comparision with first four working days of week. The situation is opposit in other districts, such as R. Total number of crime incidents decreased here at weekend.

Distribution of thefts of vechicles by days of week repeats in general terms diastribution of total crime in the districts. Number of vechicle thefts is increasing in district E to the end of week and it is decreasing in district R at the same time. However, the least number of vechicle thefts in both districts is at Thursday.

```python
pivot_by_district = seattle_data.pivot_table('RMS CDW ID', aggfunc='count', index='day_of_week', columns='District/Sector')
pivot_by_district.fillna(0, inplace=True)
pivot_by_district_vehtheft = seattle_data[seattle_data['Offense Type']=='VEH-THEFT-AUTO'].pivot_table('RMS CDW ID', aggfunc='count', index='day_of_week', columns='District/Sector')
pivot_by_district_vehtheft.fillna(0, inplace=True)
week   = ['Monday', 'Tuesday', 'Wednesday',  'Thursday',  'Friday', 'Saturday', 'Sunday']

#plt.figure()
fig, ax = plt.subplots(2, sharex=True, figsize=(10,12))
pivot_by_district[['E', 'R']].plot(kind='bar', legend='True', ax=ax[0])
ax[0].set_title('Total number of crime in districts distributed by days of week', fontdict={'fontsize': 16, 'fontweight':3})
ax[0].set(xlabel='Day of week', ylabel='Total number of crime')

#plt.subplot(122)
pivot_by_district_vehtheft[['E', 'R']].plot(kind='bar', legend=True, ax=ax[1])
ax[1].set_title('Number of vechicle thefts in districts distributed by days of week', fontdict={'fontsize': 16, 'fontweight':3})
ax[1].set(xlabel='Day of week', ylabel='Number of vechicle thefts')
plt.xticks(range(7), week, rotation='vertical')
plt.show()
```

## Data sources and methods

Crime data from Summer 2014 in Seattle are from the course [github repository](https://github.com/uwescience/datasci_course_materials/tree/master/assignment6). I got shapes of Seattle beats from [Seattle GIS Open Data](http://data-seattlecitygis.opendata.arcgis.com/datasets/36378b7acb8a464c8019b9618fecd0dd_2).

I used Python packages numpy, pandas, matplotlib, seaborn for analysis of the data. Packages geopandas and shapely were used for processing spatial distribution of crime. I grouped by beats/zones number of crime incidents and joined result with shapes of the beats/zones. After this, I divided number of crimes in beats/zones on their area and put the result on a choropleth map. I added dots with coordinates of vechicle thefts on the map.
